```python
import csv


def parsegraph(filename) :
    if filename.endswith('.json'):
        with open(filename) as f :
            G = json.load(f)
        return G


    if filename.endswith('.csv'):
        D = {}
        V = {}
        with open(filename) as f :
            doc = csv.reader(f)
            for lgn in doc :
                liste = lgn
                liste[2] = int(liste[2])
                V[liste[0]] = 1
                V[liste[1]] = 1
                if liste[0] in D :
                    D[liste[0]][liste[1]] = liste[2]
                else :
                    D[liste[0]] = {liste[1] : liste[2]}
        for v in V :
            if not v in D :
                D[v] = {}
        return D

```

```python
# ce qui donnerait pour notre graphe témoin

G = {'a': {'b': 7, 'd': 9, 'c': 14},
    'b': {'d': 10, 'e': 15},
    'c': {'d': 2, 'f': 9},
    'd': {'e': 11},
    'e': {'f': 6},
    'f' : {},}
```
