**Exercice 3**

*Atteignabilité*

```python
def reachables(graph, s) :
    R = {s}
    pile = [s]
    while pile :
        p = pile.pop()
        R.add(p)
        for f in graph[p] :
            if f not in R :
                pile.append(f)
    return R

```

```python
# pareil mais orienté test et pas debug
# on teste le résultat sur chaque sommet
reach = parsegraph("data/reach.csv")
reach_reachables = {
    'a': {'a', 'b', 'c', 'd', 'e', 'f'},
 'b': {'a', 'b', 'c', 'd', 'e', 'f'},
 'd': {'a', 'b', 'c', 'd', 'e', 'f'},
 'c': {'a', 'b', 'c', 'd', 'e', 'f'},
 'e': {'e', 'f'},
 'f': {'e', 'f'}}


for vertex, expected_reachables in reach_reachables.items():
    print(f"en partant de {vertex} → {expected_reachables == reachables(reach, vertex)}")
```
